```
from unicodedata import normalize
import re
```

```
#texto = open('la liebre.txt','r').read().lower()
texto = open('la liebre.txt').read()
#Tomado de: https://stackoverflow.com/questions/14067267/lower-case-from-a-text-file/37497605
```

```
def normalizacion_texto (texto_in):
  acentos = dict.fromkeys(map(ord, u'\u0301\u0308'), None)
  texto_2 = normalize('NFKC', normalize('NFKD', texto).translate(acentos))
  #Tomado de: https://es.stackoverflow.com/questions/135707/c%C3%B3mo-puedo-reemplazar-las-letras-con-tildes-por-las-mismas-sin-tilde-pero-no-l
  texto_3 = re.sub(r'[^\w\s]','',texto_2)
  #Tomado de: https://www.delftstack.com/es/howto/python/how-to-strip-punctuation-from-a-string-in-python/
  out = "".join(char for char in texto_3 if not char.isdigit())
  #Tomado de: https://es.stackoverflow.com/questions/348668/eliminar-numeros-de-una-cadena-de-palabras
  return out
```

```
txt_norm = normalizacion_texto(texto)
```

```
words = txt_norm.split()
' '.join(words)
```

```
words = txt_norm.split()
d = dict()

for word in words:
		if word in d:
			d[word] = d[word] + 1
		else:
			d[word] = 1

#Tomado de: https://www.geeksforgeeks.org/python-count-occurrences-of-each-word-in-given-text-file-using-dictionary/
```

```
type(txt_norm)
```

```
orden = dict(sorted(d.items(), key= lambda x: x[1], reverse=True))

print("El top 10 para las palabras de una letra es:")
for key in list(orden.keys()):
  if (len(key)== 1):
      print(key, ":", d[key], "veces")
#Tomado de: https://pybit.es/dict-ordering.html
```

```
print("El top 10 para las palabras de dos letras es:")
for key in list(orden.keys())[0:50]:
      if (len(key)== 2):
        print(key, ":", d[key], "veces")

#Tomado de: https://stackoverflow.com/questions/7971618/return-first-n-keyvalue-pairs-from-dict
```
